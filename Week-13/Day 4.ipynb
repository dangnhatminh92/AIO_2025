```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Đọc hình ảnh
img = cv2. imread ("/home/dangnhatminh92/Documents/AIO/AIO_2025/WARMING UP/Coding/Week-13/input.jpg")
img = cv2. cvtColor (img , cv2. COLOR_BGR2RGB )

# Điều chỉnh độ sáng hình ảnh
new_img = img.astype(np.float64) + 50
new_img = np.clip(new_img, 0, 255)
new_img = new_img.astype(np.uint8)

# Hiển thị hình ảnh
plt . figure ( figsize =(10 , 10))

plt . subplot (121)
plt . imshow (img)
plt . title (" Original Image ")

plt . subplot (122)
plt . imshow ( new_img )
plt . title (" Brightness Image ")
plt . show ()
```

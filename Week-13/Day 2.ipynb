```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('input.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
cv2.imwrite('saved.jpg', img)
```

```python
# Bài 1: Đọc và hiển thị ảnh

def display_image(image_path):
    img = cv2.imread(image_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.show()

display_image('input.jpg')
```

```python
# Bài 2: Lưu ảnh dưới tên khác
def save_image(image_path, output_path):
    img = cv2.imread(image_path)
    cv2.imwrite(output_path, img)

save_image('input.jpg', 'saved01.jpg')
```

```python
# Bài 3: Chuyển ảnh từ BGR sang RGB
imgrgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(imgrgb)
plt.axis('off')
plt.show()
```

```python
# Bài 4: Chuyển ảnh từ BGR sang HSV
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv)
plt.axis('off')
plt.show()
```

```python
# Bài 5: Chuyển ảnh sang Grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imwrite('gray.jpg', img_gray)
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.show()
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

combined = cv2.hconcat([img_rgb, img_hsv])


plt.imshow(combined)
plt.axis('off')
plt.show()


```
